#PYTHON BASICS

1. What is Python, and why is it popular?
  - Python is a high-level, interpreted programming language known for its simplicity and readability.Python stands out because it blends simplicity with powerful capabilities.

2. What is an interpreter in Python?
  - An interpreter is a program that reads and runs code line by line. In Python, the interpreter executes each command directly without converting the entire code into machine language first.

3. What are pre-defined keywords in Python?
  - Pre-defined keywords in Python are reserved words that have special meaning in the language. We can't use them as variable names because they’re part of Python’s core syntax.

4. Can keywords be used as variable names?
  - No

5. What is mutability in Python?
  - Mutability refers to whether an object’s content can be changed after it's created.

6. Why are lists mutable, but tuples are immutable?
  - Lists are mutable so you can dynamically manage elements and Tuples are immutable to ensure stability and hashability.

7. What is the difference between “==” and “is” operators in Python?
  - In Python, the == operator is used to compare the values of two objects and returns True if they are equal.
  - The is operator checks whether two variables refer to the exact same object in memory. Even if two objects have the same value, they might not be the same in memory, so is can return False.

8. What are logical operators in Python?
  - Logical operators are used to perform boolean logic in Python. They combine conditional statements and return True or False based on the logic applied.

9. What is type casting in Python?
  - Type casting in Python refers to converting one data type into another. It allows developers to explicitly change the type of a variable.

10. What is the difference between implicit and explicit type casting?
  - Implicit type casting is automatic—Python converts types during operations.
  - Explicit type casting is manual—you tell Python what type you want.

11.What is the purpose of conditional statements in Python?
    - Conditional statements help programs make decisions based on different conditions. They allow the flow of execution to change depending on whether certain criteria are met.

12. How does the elif statement work?
  - The elif statement stands for “else if” and is used to check multiple conditions one by one. It comes after an if statement and allows you to test another condition if the first one is false. If the elif condition is true, its block of code runs—otherwise, Python moves on.

13.  What is the difference between for and while loops?
  - The for loop is used when you know how many times you want to repeat something, usually with a sequence like a list or range.
  - The while loop is used when the number of repetitions depends on a condition being true.

14. Describe a scenario where a while loop is more suitable than a for loop?
  - A while loop is ideal when the number of iterations isn’t known in advance and depends on a changing condition. For example, if we are writing a program that asks a user to enter a password until they get it right. You don’t know how many attempts it will take, so a while loop is proper for this condition.

```python
# 6.  Write a function to demonstrate the behavior of mutable and immutable arguments.
```

```python
def modify_list(my_list):
  """Demonstrates modifying a mutable argument (list)."""
  my_list.append(6)
  print("Inside function (mutable):", my_list)

def modify_integer(my_integer):
  """Demonstrates attempting to modify an immutable argument (integer)."""
  my_integer = my_integer + 1
  print("Inside function (immutable):", my_integer)

# Demonstrate with a mutable argument (list)
my_list = [1, 2, 3, 4, 5]
print("Before function call (mutable):", my_list)
modify_list(my_list)
print("After function call (mutable):", my_list)

print("-" * 20)
# Demonstrate with an immutable argument (integer)
my_integer = 10
print("Before function call (immutable):", my_integer)
modify_integer(my_integer)
print("After function call (immutable):", my_integer)
```

```python
# 9. Write a Python program to convert user input from string to integer, float, and boolean types

# Get input from the user as a string
user_input = input("Enter a value: ")

print("\nConverted Values:")

# Convert to integer
try:
    int_value = int(user_input)
    print(f"Integer: {int_value}")
except ValueError:
    print("Integer: Conversion failed (invalid format)")

# Convert to float
try:
    float_value = float(user_input)
    print(f"Float: {float_value}")
except ValueError:
    print("Float: Conversion failed (invalid format)")

# Convert to boolean
bool_value = bool(user_input.strip())
print(f"Boolean: {bool_value}")
```

```python
#10. Write code to demonstrate type casting with list elements

# Sample list of strings representing numbers
str_list = ["1", "2", "3.5", "4.0", "True", "False", "hello"]

# Converting string list elements to different types
int_list = []
float_list = []
bool_list = []

for item in str_list:
    # Try converting to int
    try:
        int_list.append(int(float(item)))
    except ValueError:
        int_list.append(None)

    # Try converting to float
    try:
        float_list.append(float(item))
    except ValueError:
        float_list.append(None)

    # Convert to boolean (non-empty strings are True except "False")
    bool_list.append(item.lower() != "false" and bool(item.strip()))

# Display Results
print("Original List:", str_list)
print("As Integers:", int_list)
print("As Floats:", float_list)
print("As Booleans:", bool_list)
```

```python
#14. Write a program to reverse a string using a while loop
original_str = input("Enter a string: ")

reversed_str = ""
index = len(original_str) - 1

while index >= 0:
    reversed_str += original_str[index]
    index -= 1

print(f"Reversed string: {reversed_str}")
```

```python
# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop

num = int(input("Enter a non-negative integer: "))

factorial = 1
current = num

if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    while current > 1:
        factorial *= current
        current -= 1

    print(f"Factorial of {num} is {factorial}")
```
